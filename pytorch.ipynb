```python
%pip install torchvision

import torch
import torch.nn as nn
import torchvision
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
```

```python
# transforms 

transform=transforms.Compose([

    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5),(0.5,0.5,0.5))
])
```

```python
train_data=datasets.CIFAR10(root='./data',train=True,transform=transform,download=True)
test_data=datasets.CIFAR10(root='./data',train=False,transform=transform,download=True)
```

```python
train_loader = DataLoader(train_data, batch_size=64, shuffle=True)
test_loader = DataLoader(test_data, batch_size=64, shuffle=False)
```

build the architecture

```python
class CNN_Cifar10(nn.Module):
    def __init__(self):
        super().__init__()

        #feature extraction
        self.cnn1= nn.Conv2d(in_channels=3,out_channels=6,kernel_size=5)
        self.pool1=nn.MaxPool1d(2,2)
        self.cnn2= nn.Conv2d(6,16,3)
        self.pool2=nn.MaxPool2d(2,2)

        #classifier
        nn.Flatten()
        self.fc1=nn.Linear(6*6*16,4096)
        self.fc2=nn.Linear(4096,2048)
        self.fc3=nn.Linear(2048,1024)
        self.fc4=nn.Linear(1024,512)
        self.fc5=nn.Linear(512,256)
        self.fc6=nn.Linear(256,10)

    def forward(self, x):
        #feature extraction
        x=self.cnn1(x)
        x=nn.ReLU(x)
        x=self.pool1(x)
        x=self.pool2(nn.ReLU(self.cnn2(x)))
        x=nn.Flatten(x)

        #classifier
        x=self.fc1(x)
        x=nn.ReLU(x)
        x=self.fc2(x)
        x=nn.ReLU(x)
        x=self.fc3(x)
        x=nn.ReLU(x)
        x=self.fc4(x)
        x=nn.ReLU(x)
        x=self.fc5(x)
        x=nn.ReLU(x)
        x=self.fc6(x)

        return x
```

```python
model=CNN_Cifar10()
criterion=nn.CrossEntropyLoss()
optimizer=torch.optim.Adam(model.parameters(), lr=0.0001)
```

Train

```python
for epoch in range(10):
    for images,labels in enumerate(train_loader):
        outputs=model(images)
        loss=criterion(outputs,labels)
        loss.backward()
        optimizer.zero_grad()
        optimizer.step()
```
